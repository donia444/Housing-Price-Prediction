```python
df.hist(bins=50,figsize=(20,15),color='purple')
plt.show()
```

```python

plt.figure(figsize=(10,8))
sns.histplot(df['median_house_value'],color='purple',kde=True,bins=50)
plt.xlabel('Median house val')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=df,x='median_income',y='median_house_value',color='purple')
plt.show()
```

```python
df_corr=df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(df_corr,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
